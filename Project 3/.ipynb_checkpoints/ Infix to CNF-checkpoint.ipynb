# Laws of Propositional Logic


```python
""" BIDIRECTIONAL ELIMINATION"""
def bidirectional_elimination(sentence): 
    if(sentence[0]=="iff"):
        literal_1=sentence[1]
        literal_2=sentence[2]
        sentence[0]="and"
        sentence[1]=["implies",literal_1,literal_2]
        sentence[2]=["implies",literal_2,literal_1]

    for literal in sentence:
        if(len(literal)>1):
            bidirectional_elimination(literal)
```

```python
""" IMPLICATION ELIMINATION"""
def implication_elimination(sentence):  
    if(sentence[0]=="implies"):
        literal_1=sentence[1]
        sentence[0]="or"
        sentence[1]=["not",literal_1]

    for literal in sentence:
        if(len(literal)>1):
            implication_elimination(literal)
```

```python
""" DE-MORGAN'S LAW """
def demorgans(sentence): 
    if(sentence[0]=="not"):
        literal=sentence[1]

        if(literal[0]=="not"):
            del sentence[:]
            if(len(literal[1])==1):
                sentence.append(literal[1])
            else:
                for literal_1 in literal[1]:
                    sentence.append(literal_1)

                if(len(sentence)>1):
                    demorgans(sentence)

        elif(literal[0]=="or"):
            del sentence[:]
            sentence.append("and")
            for literal_1 in literal:
                if literal_1== "or":
                    continue
                else:
                    sentence.append(["not",literal_1])

        elif(literal[0]=="and"):
            del sentence[:]
            sentence.append("or")
            for literal_1 in literal:
                if literal_1== "and":
                    continue
                else:
                    sentence.append(["not",literal_1])
        
       

    for literal in sentence:
        if(len(literal)>1 ):
            demorgans(literal)
```

```python
""" DISTRIBUTIVE LAW """
def distributivity(sentence):
    if(sentence[0]=="or"):
        if(len(sentence[2])>1  and sentence[2][0]=="and"):
            literal_1=sentence[1]
            literal_2=sentence[2]
            del sentence[:]
            sentence.append("and")
            for literal_x in literal_2:
                if literal_x=="and":
                    continue
                sentence.append(["or",literal_1,literal_x])
        elif(len(sentence[1])>1  and sentence[1][0]=="and"):
            literal_1=sentence[1]
            literal_2=sentence[2]
            del sentence[:]
            sentence.append("and")
            for literal_x in literal_1:
                if literal_x=="and":
                    continue
                sentence.append(["or",literal_x,literal_2])

    for literal in sentence:
        if(len(literal)>1 ):
            distributivity(literal)
```

```python
""" OUTER RECURSIVE DISTRIBUTIVE LAW"""
def distributivity_recursion(sentence):
    if(sentence[0]=="or"):
        if(len(sentence[2])>1  and sentence[2][0]=="and"):
            literal_1=sentence[1]
            literal_2=sentence[2]
            del sentence[:]
            sentence.append("and")
            for literal_x in literal_2:
                if literal_x=="and":
                    continue
                sentence.append(["or",literal_1,literal_x])
        elif(len(sentence[1])>1  and sentence[1][0]=="and"):
            literal_1=sentence[1]
            literal_2=sentence[2]
            del sentence[:]
            sentence.append("and")
            for literal_x in literal_1:
                if literal_x=="and":
                    continue
                sentence.append(["or",literal_x,literal_2])
```

```python
""" ASSOCIATIVE LAW """
def associvity(sentence):
    if(sentence[0]=="and"):
        temp_sentence=copy.deepcopy(sentence)
        del sentence[:]
        for literal in temp_sentence:
            if literal=="and":
                sentence.append("and")
            else:
                if literal[0]!="and":
                    sentence.append(literal)
                else:
                    for literal_x in literal:
                        if literal_x=="and":
                            continue
                        else:
                            sentence.append(literal_x)

    if(sentence[0]=="or"):
        temp_sentence=copy.deepcopy(sentence)
        del sentence[:]
        for literal in temp_sentence:
            if literal=="or":
                sentence.append("or")
            else:
                if literal[0]!="or":
                    sentence.append(literal)
                else:
                    for literal_x in literal:
                        if literal_x=="or":
                            continue
                        else:
                            sentence.append(literal_x)

    for literal in sentence:
        if(len(literal)>1):
            associvity(literal) 
```

```python
""" REMOVE PARENS """   
def remove_parens(sentence):
     i=0
     for literal in sentence:
        if(len(literal)==1):
            temp=str(literal)  
            if(len(temp)>1):
              sentence[i]=temp[2]
        elif(len(literal)>1):
            remove_parens(literal)
        i=i+1
```

```python
""" REMOVE DUPLICATES """
def remove_duplicates(sentence):
    delete_literals=[]
    if(sentence[0]=="and" or sentence[0]=="or"):
        for x in range(0,len(sentence)):
            for y in range(x+1,len(sentence)):
                literal_1=sorted(sentence[x])
                literal_2=sorted(sentence[y])
                if literal_1==literal_2:
                   delete_literals.append(sentence[y])
 
    if(len(delete_literals)>=1):
        temp_sentence=[]
        for literal in delete_literals:
            flag=1
            for literal_x in sentence:
                if (literal==literal_x and flag==1):
                    flag=0
                    continue
                else:
                    temp_sentence.append(literal_x)
        
        del sentence[:]
        for literal in temp_sentence:
            sentence.append(literal)
        

    for literal in sentence:
        if(len(literal)>1):
            remove_duplicates(literal) 
```

```python
""" REMOVE SINGLE LITERALS WITH AND-OR """
def remove_single_literals(sentence):
    if(len(sentence)==2 and sentence[0]!="not"):
        del sentence[0]

    for literal in sentence:
        if(len(literal)>1 and isinstance(literal, list)):
            remove_single_literals(literal)
```

```python
""" QUALITY CHECK """
def quality_check(sentence):
    if(len(sentence)==1 and len(sentence[0])>1):
        literal=sentence[0]
        del sentence[:]
        for literal_x in literal:
            sentence.append(literal_x)

    for literal in sentence:
        if(len(literal)>1):
            quality_check(literal)
```

```python
""" CONVERT TO CNF """
def convert_CNF(sentence):
    bidirectional_elimination(sentence)
    implication_elimination(sentence)
    demorgans(sentence)
    distributivity(sentence)  
    flag=1
    while(sentence[0]=="or" and flag==1):
        if (len(sentence[2])>1  and sentence[2][0]=="and"):
            distributivity_recursion(sentence)
            flag=1
        elif (len(sentence[1])>1  and sentence[1][0]=="and"):
            distributivity_recursion(sentence)
            flag=1
        else:
            flag=0

    associvity(sentence)
    remove_parens(sentence)
    remove_single_literals(sentence)
    quality_check(sentence)
    remove_parens(sentence)

    check=1
    if(sentence[0]=="and"):
       temp_literal=sentence[1]
       for x in range(1,len(sentence)):
           if temp_literal != sentence[x]:
               check=0

    if check==1:
        remove_duplicates(sentence)
        remove_single_literals(sentence)
        quality_check(sentence)
        remove_parens(sentence)
    return sentence
```
